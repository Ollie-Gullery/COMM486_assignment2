```python
# Defining Variables and list
ticker_list = ['AAPL', 'TSLA', 'META', 'NVDA', 'MSFT']

# define your start and end date
start = datetime.datetime(2023, 1, 1)
end = datetime.datetime(2023, 12, 31)
```

```python
# Method for question 1
def question_1(ticker, start_date, end_date):
    yahoo_data = yf.download(ticker, start=start, end=end)
    adj_price = yahoo_data['Adj Close']
    returns  = adj_price.pct_change()
    ret_mean = returns.mean()
    ret_std = returns.std()
    annualized_ret_mean=((1+ret_mean)**250)-1
    annualized_ret_std=ret_std*(250**(1/2))
    print(f'Return Data for {ticker} for the year {start_date.year} \n'
      f'Mean Return (%): {100*np.mean(returns):.8f}% \n'
      f'Standard Deviation (%): {100*np.std(returns):.8f}% ')
    print('\n')
    print(
    f'Annualized Return Data for {ticker} for the year {start_date.year}\n'
    f'Mean Annualized Return (%): {annualized_ret_mean*100:.8f}%\n'
    f'Annualized Standard Deviation (%): {annualized_ret_std*100:.8f}%'
    )
    print('\n')
    
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
    
    # Plot Adjusted Close Price
    ax1.plot(yahoo_data.index, adj_price, label='Adj Close')
    ax1.set_title(f'{ticker}: Adjusted Close Prices Over Time')
    ax1.set_xlabel('Date')
    ax1.set_ylabel('Adjusted Close Price')
    ax1.tick_params(axis='x', rotation=45)
    ax1.legend()
    ax1.grid(True)
    
    # Plot Returns
    ax2.plot(yahoo_data.index, returns, label='Returns')
    ax2.set_title(f'{ticker}: Returns Over Time')
    ax2.set_xlabel('Date')
    ax2.set_ylabel('Returns')
    ax2.tick_params(axis='x', rotation=45)
    ax2.legend()
    ax2.grid(True)
    
    plt.tight_layout()
    plt.show()
```

### Question 2.1. Why does code use 250 to annualize daily returns?
This is because the number of trading days per year is 250, which is different from a calendar year which has 365 days.

```python
filename = "TSX60_Listing_2023.csv"
TSX60Companies= pd.read_csv(filename)
ticker = list(TSX60Companies['YAHOO_TICKER'])
yahoo_data = yf.download(ticker,start=start, end=end)

yahoo_data.head(1)
```

```python
adj_price = yahoo_data['Adj Close']

# show adj_price
adj_price

# calculate simple returns
returns = adj_price.pct_change()

#mean
ret_mean=returns.mean()
annualized_ret_mean=((1+ret_mean)**250)-1
#annualized mean
print(annualized_ret_mean.head(2))

#standard Deviation
ret_std=returns.std()
annualized_ret_std=ret_std*(250**(1/2))
#show standard deviation and annualized standard deviation
annualized_ret_std.head(2)
```

```python
# parameters for the efficient Frontier, feel free to adjust num_portfolios and the risk free rate to observe changes
mean_returns=ret_mean
cov_matrix = returns.cov()
num_portfolios = 25000  # try other number of portfolios 100k, 50k, 10k, 1000, 100
risk_free_rate = 0.04

#annualize the portfolio returns
def portfolio_annualized_performance(weights, mean_returns, cov_matrix):
    returns = np.sum(mean_returns*weights ) *250
    std = np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights))) * np.sqrt(250)
    return std, returns

# build random portfolios out of all the stocks
def random_portfolios(num_portfolios, mean_returns, cov_matrix, risk_free_rate):
    results = np.zeros((3,num_portfolios))
    weights_record = []
    for i in range(num_portfolios):
        weights = np.random.random(60)
        weights /= np.sum(weights)
        weights_record.append(weights)
        portfolio_std_dev, portfolio_return = portfolio_annualized_performance(weights, mean_returns, cov_matrix)
        results[0,i] = portfolio_std_dev
        results[1,i] = portfolio_return
        results[2,i] = (portfolio_return - risk_free_rate) / portfolio_std_dev
    return results, weights_record
```

### Question 2.6. Differences between the tables

Analyzing the two tables for the Maximum Sharpe Ratio Portfolio Allocation and the Minimum Volatility Portfolio Allocation reveals distinct strategic outcomes. The Maximum Sharpe Ratio Portfolio consistently shows higher annualized returns across various portfolio sizes, ranging from 0.15 to 0.19, indicating a strategy focused on maximizing returns for a given level of risk. However, this strategy also incurs higher annualized volatility, indicated by a volatility ranging from 0.11 to 0.19, reflecting the increased risk taken. In contrast, the Minimum Volatility Portfolio demonstrates lower returns, from 0.11 to 0.14, with a reduced volatility of 0.1 across the board, emphasizing the focus on risk minimization. The largest and smallest holdings by amount in each portfolio also suggest a diversification strategy that differs between the two approaches, with the Maximum Sharpe Ratio Portfolio having a narrower range between its largest and smallest holdings compared to the Minimum Volatility Portfolio, which shows a broader dispersion, reflecting a more conservative allocation with a potential emphasis on stability over high returns.

![Minimum Volatility](minimumvol.png)

```python
num_portfolio_list = [100000, 50000, 25000, 1000, 100]
def question_2_num_portfolio(num_portfolio_list, mean_returns=mean_returns, cov_matrix=cov_matrix, risk_free_rate=risk_free_rate):
    for num_portfolios in num_portfolio_list:
      results, weights = random_portfolios(num_portfolios,mean_returns, \
                                           cov_matrix, risk_free_rate)
      max_sharpe_idx = np.argmax(results[2])
      sdp, rp = results[0,max_sharpe_idx], results[1,max_sharpe_idx]
      max_sharpe_allocation = pd.DataFrame(weights[max_sharpe_idx],\
                                           index=adj_price.columns,columns=['allocation'])
      max_sharpe_allocation.allocation = [round(i*100,2)for i in max_sharpe_allocation.allocation]
      max_sharpe_allocation = max_sharpe_allocation

      min_vol_idx = np.argmin(results[0])
      sdp_min, rp_min = results[0,min_vol_idx], results[1,min_vol_idx]
      min_vol_allocation = pd.DataFrame(weights[min_vol_idx],index=adj_price.columns,columns=['allocation'])
      min_vol_allocation.allocation = [round(i*100,2)for i in min_vol_allocation.allocation]
      min_vol_allocation = min_vol_allocation

      largest_ticker_vol = min_vol_allocation['allocation'].idxmax()
      smallest_ticker_vol = min_vol_allocation['allocation'].idxmin()
      largest_weight_vol = min_vol_allocation.loc[largest_ticker_vol, 'allocation']
      smallest_weight_vol = min_vol_allocation.loc[smallest_ticker_vol, 'allocation']

      largest_ticker_sharpe = max_sharpe_allocation['allocation'].idxmax()
      smallest_ticker_sharpe = max_sharpe_allocation['allocation'].idxmin()
      largest_weight_sharpe = max_sharpe_allocation.loc[largest_ticker_sharpe, 'allocation']
      smallest_weight_sharpe = max_sharpe_allocation.loc[smallest_ticker_sharpe, 'allocation']

      print(f'Results for Portolio (Num: {num_portfolios})')
      print("-"*80)
      print("Maximum Sharpe Ratio Portfolio Allocation\n")
      print("Annualised Return:", round(rp,2))
      print("Annualised Volatility:", round(sdp,2))
      print("\n")
      print(f'Largest Portfolio Holding ({num_portfolios}): {largest_ticker_sharpe}: {largest_weight_sharpe} \nSmallest Portfolio Holding ({num_portfolios}): {smallest_ticker_sharpe}: {smallest_weight_sharpe} ')
      print("-"*80)
      print("Minimum Volatility Portfolio Allocation\n")
      print("Annualized Return:", round(rp_min,2))
      print("Annualized Volatility:", round(sdp_min,2))
      print("\n")
      print(f'Largest Portfolio Holding ({num_portfolios}): {largest_ticker_vol}: {largest_weight_vol}  \nSmallest Portfolio Holding ({num_portfolios}): {smallest_ticker_vol}: {smallest_weight_vol}')
      print("\n")
```

**Why Might EBITDA Yield be more relevant for retailers?**

EBITDA can be more relevant for retailers for several reasons:

- **Focus on Operations**: EBITDA focuses on the earnings generated from the core operations of a business and excludes non-operational factors (such as interest, taxes, depreciation, and amortization). For retailers, whose primary source of revenue comes from selling good, this metric will thus provide a clearer picture of operational performance. Additionally, since EBITDA excludes non-operating expenses, it allows for a more straightforward comparison of the operational performance of different retails regardless of capital structure or differences in accounting methods. 
- **Capital Intensive Nature**: Retail businesses often require significant investments in inventory, property, and equipment. Thus, by excluding depreciation and amortization, EBITDA provides a better reflection of the cash flows available to cover these capital investments.
- **Growth Evaluation / Valuation**:Investors often use EBITDA yield as a measure of efficiency and profitability, as they indicate a retailer's potential for growth. It provides insights into a company's ability to generate returns on invested capital, which means its valuable for assessing long-term performance and investment opportunities. Hence why EBITDA is commonly used in valuation models during M&A within the retail sector.  

Ultimately, book yield is suitable for evaluating banks due to their distinct financial characteristics, but EBITDA yield is preferred for retailers as it aligns more closely with the operational and financial dynamics of the retail industry. Ultimately, providing a better understanding of their profitability and performance.

```python
ctc_data = pd.read_csv("ctc_data.csv")

ctc_data.head(10)
```

```python
def calc_one_month_return(ticker,start_date,end_date):

    yahoo_data = yf.download(ticker, start=start_date, end=end_date)
    adj_price = yahoo_data['Adj Close']

    start_price = adj_price[0]
    end_price = adj_price[-1]

    one_month_return = (end_price - start_price) / start_price
    
    return one_month_return
```

```python
# Defining Variables and list
start_date = datetime.datetime(2024, 1, 1)
end_date = datetime.datetime(2024, 2, 1)

list_of_tickers = ctc_data['Ticker '].head(10)
earnings_yield = ctc_data['Earnings Yield (1/ BF P/E)'].head(10)
ebitda_yield = ctc_data['EBITDA Yield (1/EV/EVITDA)'].head(10)
one_month_return = [calc_one_month_return(ticker,start_date,end_date) for ticker in list_of_tickers]
```

```python
import statsmodels.api as sm

def run_regression(earnings_yield, ebitda_yield, one_month_return):
    earnings_yield_z = (earnings_yield - np.mean(earnings_yield)) / np.std(earnings_yield)
    ebitda_yield_z = (ebitda_yield - np.mean(ebitda_yield)) / np.std(ebitda_yield)
    
    # Prepare the data for regression
    X = sm.add_constant(np.column_stack((earnings_yield_z, ebitda_yield_z))) 
    y = one_month_return
    
    # Run the regression
    model = sm.OLS(y, X).fit()
    
    print(f"R-squared: {model.rsquared:.6f}")
    print(f"Adjusted R-squared: {model.rsquared_adj:.6f}")
    print(f"Intercept Coefficient: {model.params[0]:.6f}, t-stat: {model.tvalues[0]:.6f}")
    print(f"Earnings Yield Coefficient: {model.params[1]:.6f}, t-stat: {model.tvalues[1]:.6f}")
    print(f"EBITDA Yield Coefficient: {model.params[2]:.6f}, t-stat: {model.tvalues[2]:.6f}")

    return 
```

```python
run_regression(earnings_yield, ebitda_yield, one_month_return)
```

## Analysis of Results

The regression results indicate that our model has a moderate explanatory power with an R-squared of **0.545858**. This indicates that *54.59%* of the variation in our model can be explained with our predictor variables. The adjusted R-squared of **0.416103** accounts for the number of predictors and the sample size, offering a slightly more conservative assessment - likely due to our small sample size. The negative intercept coefficient of **-0.049871** with a t-stat of **-1.928537** suggests a slight negative baseline effect. The earnings yield coefficient is also negative **-0.066920** and more significantly influences the model, as indicated by its t-stat of **-2.568491**. Conversely, the EBITDA yield coefficient exhibited a positive relationship (**0.026692**) with our dependent variable one month returns, albeit with a lower impact (-stat: **1.024493**), suggesting a weaker contribution to the model.

I should note that while our R-squared of 54.59% indicated a moderate relationship, there may be a more complex non-linear relationship that our model failed to capture which may have better mapped the relationship between the independent and dependent variables. 

```python
def run_composite_regression(earnings_yield, ebitda_yield, one_month_return):
    # calculating composite value & z score
    composite_value = 0.5 * earnings_yield + 0.5 * ebitda_yield
    composite_value_z = (composite_value - np.mean(composite_value)) / np.std(composite_value)
    
    X = sm.add_constant(composite_value_z)  
    y = one_month_return
    # creating regression model
    model = sm.OLS(y, X).fit()

    # formatting outputs 
    print(f"R-squared: {model.rsquared:.6f}")
    print(f"Adjusted R-squared: {model.rsquared_adj:.6f}")
    print(f"Intercept Coefficient: {model.params['const']:.6f}, t-stat: {model.tvalues['const']:.6f}")
    print(f"Composite Value Coefficient: {model.params[0]:.6f}, t-stat: {model.tvalues[0]:.6f}")

    return
```
